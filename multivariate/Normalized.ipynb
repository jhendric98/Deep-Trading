```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
def compute_daily_returns(df):
    """Compute and return the daily return values."""
    df1 = df.copy()
    df1[1:] = (df[1:]/df[:-1].values) -1
    df1.iloc[0, : ] = 0
    return df1
```

```python
df = pd.read_csv('./data/AAPL1216.csv', index_col="Date")
df.head(3)
```

```python
df_daily_rtns = compute_daily_returns(df)
df_daily_rtns.head(3)
```

```python
openp = df_daily_rtns['Open'].tolist()
highp = df_daily_rtns['High'].tolist()
lowp = df_daily_rtns['Low'].tolist()
closep = df_daily_rtns['Adj Close'].tolist()
volumep = df_daily_rtns['Volume'].tolist()
```

```python
df_daily_rtns["Current"] = df[["Close"]].values
df_daily_rtns["Future"] = df[["Close"]].shift(-30).values
df_daily_rtns["Forecast"] = (df[["Close"]].shift(-30).values / df[["Close"]].values) - 1
df_daily_rtns.head(40)
```
